```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
df = pd.read_csv('C:\\Users\\Arseniy\\Downloads\\archive (3)\\movies.csv')
import scipy.stats as stats
```

```python
#checking data on NaN values
df.info()
df.isnull().count()
```

```python
#remove values Not A Number (NaN)
df = df.dropna() 
```

```python
#change datatype of column released, making new column yearcorrect
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
#scatterplot for budget and gross values
plt.scatter(x=df.budget, y=df.gross);
plt.title('Budget VS Gross Earnings');
plt.xlabel('Budget for Film');
plt.ylabel('Gross Earnings');
```

```python
#observing positive correlation regplot
sns.regplot(x=df.budget, y=df.gross, scatter_kws={'color':'red'},line_kws={'color':'blue'});
plt.title('Budget VS Gross Earnings');
plt.xlabel('Budget for Film');
plt.ylabel('Gross Earnings');
```

```python
df.corr(method='pearson') #pearson,spearman,kendall 
```

```python
corr_matrix = df.corr(method='pearson')
sns.heatmap(data = corr_matrix, annot = True);
plt.title('Numeric Features Correlation Matrix');
```

```python
#numeric view of table
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
corr_matrix = df_numerized.corr(method='pearson')
sns.heatmap(data = corr_matrix, annot = True);
plt.title('All Features Correlation Matrix');
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs.head()
```

```python
#perform t-test
t_statistic, p_value = stats.ttest_ind(df.budget, df.gross)
# Print results
print("T-statistic: ", t_statistic)
print("P-value: ", p_value)

# Interpret the results
if p_value < 0.05:  # Assuming a significance level of 0.05
    print("Reject null hypothesis: Groups are statistically significantly different.")
else:
    print("Fail to reject null hypothesis: Groups are not statistically significantly different.")
```
